```python
import pandas as pd

df = pd.read_excel("CBPOne 2024(Responses) (Attorney privelege doc).xlsx")

df = df.dropna(subset=['Nombres y Apellidos']) #drop all rows that don't have a name/last name for client
```

```python
df['Total Family Members'] = df['Número de Mujeres que viajan (incluyendo al titular)'] + df['Número de Hombres que viajan   (incluyendo al titular)'] + df['Número de menores que viajan']
```

```python
import numpy as np
import dataframe_image as dfi
repeated_data = np.repeat(df.values, df['Total Family Members'], axis=0)

# Create a new DataFrame with the repeated rows
df_repeated = pd.DataFrame(repeated_data, columns=df.columns)

df_repeated['Nacionalidad'].value_counts(normalize=True) * 100
```
